```python
import numpy as np 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = df.groupby('id').mean().reset_index()
```

```python
df
```

```python
df.drop('id',axis=1,inplace=True)
df
```

```python
df.corr()
```

# Remove Multicolinear columns

```python
columns = list(df.columns)
for col in columns:
    if col == 'class':
        continue

    filterd_columns = [col]
    for col1 in df.columns:
        if col == col1 or col=='class':
            continue
        if df[col].corr(df[col1]) > 0.7:
            columns.remove(col1)
            continue
        else:
            filterd_columns.append(col1)
    df = df[filterd_columns]        
df.shape
```

```python
df
```

```python
df.describe()
```

```python
X = df.drop('class',axis=1)
Y = df['class']
```

```python
X.head()
```

```python
Y.describe()
```

```python
X.describe()
```

# Normalizing the dataset

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
X_normalized = MinMaxScaler().fit_transform(X)
```

```python
X_normalized
```

```python
X_normalized = pd.DataFrame(X_normalized, columns=X.columns, index=X.index)
```

```python
X_normalized
```

```python
X_normalized.describe()
```

# Selecting most influential features using  'mutual_info_classif'

```python
from sklearn.feature_selection import SelectKBest,mutual_info_classif

selector =  SelectKBest(mutual_info_classif, k =30)
```

```python
selector
```

```python
selected_columns =  selector.fit(X_normalized,Y)
selected_features = X.columns[selector.get_support()]
selected_features
```

```python
selected_features.shape
```

```python
X_selected = X_normalized[selected_features]
```

```python
X_selected.head()
```

# Splitting dataset into tran and test sets

```python
from sklearn.model_selection import train_test_split
X_train, X_val,Y_train, Y_val = train_test_split(X_selected, Y, test_size=0.2, random_state=10)
```

# Class Imbalance fixing

```python
num = Y.value_counts()
num
```

```python
sns.barplot(x=num.index, y=num.values)
plt.title("Class Distribution")
plt.xlabel("Class")
plt.ylabel("Count")
plt.show()
```

```python
from imblearn.over_sampling import RandomOverSampler
ros = RandomOverSampler(sampling_strategy=1.0,random_state=0)
X_train_resampled, Y_train_resampled = ros.fit_resample(X_train, Y_train)
```

```python
print(Y_train_resampled.value_counts())
```

```python
sns.barplot(x=Y_train_resampled.value_counts().index, y=Y_train_resampled.value_counts().values)
plt.title("Class Distribution")
plt.xlabel("Class")
plt.ylabel("Count")
plt.show()
```

# Save preprocessed data

```python
X_train_resampled.to_csv('../data/X_train_resampled.csv', index=False)
X_val.to_csv('../data/X_val.csv', index=False)
Y_train_resampled.to_csv('../data/Y_train_resampled.csv', index=False)
Y_val.to_csv('../data/Y_val.csv', index=False)
```
